```python
# Use of break statement inside loop

for val in "string":
    if val == "r":
        break
    print(val)
print("The end")
```

```python
# Program to show the use of continue statement inside loops

for val in "string":
    if val == "r":
        continue
    print(val)

print("The end")
```
